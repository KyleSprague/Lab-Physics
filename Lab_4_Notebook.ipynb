We make use of a torsion pendulum with the goal of measuring the gravitational constant $G$. Our pendulum has two point masses (m) where each point mass is a distance $l$ away from the center of the pendulum. The large masses that exert a gravitational force on the point masses are at a distance $H$. We say that the torque on the pendulum exerted by a large mass is equal to and opposite the torque from the torsion fiber. If the pendulum is in a setted state where each point mass makes an angle $\theta_0$ with the center of the disk, then the torque for the string can be expressed as 

$$\tau = k\theta_0$$  

where k is the rotational spring constant. The torque arising from the gravitational force associated with the test mass is 

$$\tau = lF2$$ 

where the factor of 2 arises from the cortribution from each mass. Setting the torques equal to each other, we have, for 

$$F = \frac{GMm}{r^2}$$ 

$$\frac{2lGMr}{r^2} = k\theta_0.$$  

Note that r is equivalent to the distance between m and M and thus the distance between the small mass m and it's equilibrium position is then H-r. This H-r component forms a triangle with $l$. Looking at this triangle, we see that  

$$sin(\theta_0) = \frac{H-r}{l}$$ 
$$lsin(\theta_0) - H = r.$$ 

Plugging this back into equation 1, we have 

$$\frac{2lGMm}{(H-lsin(\theta_0))^2} = k\theta_0$$

which we rearrange to become 

$$G = \frac{k\theta_0(H-lsin(\theta_0))^2}{2lMm}.$$

Recall that the moment of inertia for two point masses is 

$$I = 2ml^2.$$  

We can substitute in for k by recalling that a damped harmonic oscillator such as the torsion pendulum wsill have its motion described by a second order linear homogeneous differential equation assuming no external forces are pressent. This ODE can be written as 

$$I\ddot{\theta} + C\dot{\theta} + k\theta = 0$$ 

where C is the damping constant and I is the moment of inertia. Assuming C = 0 (for now), we construct a characteristic polynomial 

$$Ir^2 + k = 0$$  

with roots

$$r = \pm \sqrt{\frac{k}{I}}.$$  

Because our roots are complex, our general colution takes the form 

$$\theta(t) = Ae^{-\alpha t} cos(\beta t)$$ 

where $$\alpha = 0$$ and $$\beta = \sqrt{\frac{k}{I}}.$$ Note the similarity between this result and the expected form of the solution 

$$\theta = Ae^{\alpha t} cos(\omega_0 t),$$ Based on this similarity, we say that 

$$\omega_0 = \sqrt{\frac{k}{I}}.$$ 

Rearranging, we obtain 

$$k = w_0^2 I$$. Plugging I into equation 6, we obtain  

$$k = \omega_0^2 2ml^2$$ 

which we can plug back into equation 2 to arrive at 


$$ G = \frac{\omega_0^2 2ml^2 \theta_0 (H-lsin(\theta_0))^2}{2mlM}$$ 

$$ = \frac{\omega_0^2 l \omega_0 (H-lsin(\theta_0))^2}{M}$$

Because our system is damped due to friction with the air, it is unrealistic to have a natural frequency $$\omega_0$$ in our final result for G. To account for this, we need to solve for $$\omega$$. To do so, we plug in a a trial solution of $$\theta = Ae^{-\alpha t} cos(\omega t)$$ into the equation for the damped harmonic oscillator, which implies taking the derivative of the trial solution twice to obtain $$\dot{\theta}$$ and $$\ddot{\theta}$$. 

$$\dot{\theta} = \frac{\partial}{\partial t}[(Ae^{-\alpha t}]cos(\omega t) +
\frac{\partial}{\partial t}[cos(\omega t)](Ae^{-\alpha t})$$

$$= -\alpha Ae^{-\alpha t} cos(\omega t) - \omega Ae^{-\alpha t} sin(wt)$$ 

$$ \ddot{\theta} = \left[  \frac{\partial}{\partial t} (-\alpha Ae^{-\alpha t}) cos(\omega t) - \frac{\partial}{\partial t}(cos(\omega t)) \alpha Ae^{-\alpha t} \right] + \left[  \frac{\partial}{\partial t} (- \omega Ae^{- \alpha t} sin(\omega t) - \frac{\partial}{\partial t}(sin(\omega t)) \omega Ae^{-\alpha t}) \right ] $$

$$ = Ae^{-\alpha t}cos(\omega t) [\alpha^2 - \omega^2] + Ae^{-\alpha t} sin(\omega t) [2 \alpha \omega]$$. 

Now that we have our derivatives, we plug them back into equation 4 and similify to obtain 

$$cos(\omega t)[I(\alpha^2 - \omega^2) - c\alpha + k] + sin(\omega t)[I2\alpha \omega - C \omega] = 0.$$  

Since each respective component of equation 8 must equal zero, we obtain a system of equations: 

$$ I[\alpha^2 -\omega^2] - C\alpha + k =0$$ 
$$ I2\alpha \omega - C \omega = 0$$ 

Using equation  9, we solve for \omega^2. 

$$I \alpha^2 - \omega^2 I - c\alpha + k = 0$$ 
$$\omega^2 = \alpha^2 - \frac{C \alpha}{I} + \frac{k}{I}.$$ 

At this point, it becomes useful to solve for C using equation 10. The result of this is 

$$C = 2 I \alpha.$$ Plugging this into our current expression \omega^2 yields 

$$\omega^2 = \alpha^2 - \frac{\alpha 2 I \alpha}{I} + \frac{k}{I}$$ 

$$\omega^2 = -\alpha^2 + \omega_0^2.$$  

$$\omega_0^2 = \omega^2 + \alpha^2$$

Now if we substitute our value for \omega_0 back into equation 7, we get 

$$G = \frac{(\omega^2 + \alpha^2) l \theta_0 (H-lsin(\theta_0))^2}{M}.$$ 

Angular frequency is $$\frac{2 \pi}{T}$$ where T is the period of the oscillation. This implies that the above can be rewrritten as: 

$$G = \frac{(\frac{4 \pi^2}{T^2} + \alpha^2) l \theta_0 (H-l\theta_0)^2}{M}$$  

where we have used the small ange approximation to say that $$sin\theta_0 \approx \theta_0$$. Since we can not solve for $$\theta_0$$, we approximate its value by using a diagram where $$\theta = 4 \theta_0$$. d is the distance between equilibrium points that the sensor reaches when oscillation of the point masses becomes so small as to be undetected. Using L and D where $$\theta = \frac{d}{L}$$, we have for $$\theta_0$$ 

$$\theta_0 = \frac{d}{4L}$$ which we can substitute in to get our final result for G of 

$$G = (\frac{4 \pi^2}{T^2} + \alpha^2)(\frac{d l}{4 L M})(H - \frac{l d}{4 L})^2$$

##T Uncertainty: 

Recall 

$$T = \frac{2\pi}{\omega}$$ 

which implies that 

\begin{equation} 
\sigma_{T} = \sqrt{ \left( \frac{\partial{T}}{\partial{w}} \right)^2 \sigma_{w}^2} 
\end{equation}  

$$ = \sqrt{\left( \frac{-2 \pi}{\omega^2} \right)^2 \sigma_{w}^2} $$

where $\sigma_{\omega}$ and $\omega$ comes from the model. This means we will have a $\sigma_T$ for each trial. 

IGNORE BELOW...

$$\bar{\omega} = \frac{\sum_{n=1}^{3} \frac{\omega_i}{\sigma_i^2}}{\sum_{n=1}^{3} \frac{1}{\sigma_i^2}}  $$ 

Then to get $$\sigma_{\omega}$$, we use 

$$\sigma_{\omega} = \sqrt{\frac{1}{\sum_{n=1}^{3} \frac{1}{\sigma_i^2} }}$$ where the reason our sums have upper bounds at three is because we are only using three trials for our data. 






```python

#VALUES FROM THE MODELS 

#model 1: 
model_1_w = 0.00969809 
model_1_w_uncertainty = 4.4017*(10**-6)  
model_1_alpha = 8.6608 *(10**-4)
model_1_alpha_uncertainty = 4.4453*(10**-6)

#model 2 
model_2_w = 0.00963832 
model_2_w_uncertainty = 1.2745*(10**-5)  
model_2_alpha = 8.5159 *(10**-4)
model_2_alpha_uncertainty = 1.1194*(10**-5)

#model 3

model_3_w = 0.00936973 
model_3_w_uncertainty = 9.92663*(10**-6)  
model_3_alpha = 7.9431 *(10**-4)
model_3_alpha_uncertainty = 8.2714*(10**-6) 

def sigma_T(omega_val, sigma_omega):  
  numerator = ((-2*np.pi)**2)*(sigma_omega**2)
  denominator = omega_val**4 
  result = np.sqrt(numerator/denominator) 
  return result  

def T(omega_val): 
  return (2*np.pi)/(omega_val)

model_1_T = T(model_1_w) 
model_1_sigma_T = sigma_T(model_1_w, model_1_w_uncertainty)  
print(model_1_sigma_T)

model_2_T = T(model_2_w)
model_2_sigma_T = sigma_T(model_2_w, model_2_w_uncertainty) 
print(model_2_sigma_T)

model_3_T = T(model_3_w)
model_3_sigma_T = sigma_T(model_3_w, model_3_w_uncertainty)
print(model_3_sigma_T)

```

```python
def calculating_G(alpha, T, l, L, d, H, M): 
  first_component = ((4*(np.pi**2))/(T**2)) + alpha**2 
  second_component = (d*l)/(4*L*M) 
  third_component = (H - ((l*d)/(4*L)))**2 
  G = first_component*second_component*third_component 
  return G

G_1 = calculating_G(model_1_alpha, model_1_T, l, L, d_best_value, H, M)  
print(G_1) 
G_2 = calculating_G(model_2_alpha, model_2_T, l, L, d_best_value, H, M)  
print(G_2) 
G_3 = calculating_G(model_3_alpha, model_3_T, l, L, d_best_value, H, M)  
print(G_3)
```

#Calculating Uncertainty in G

To calculate the uncertainty in G, we will need to use an error propagation as each of the constituent components in the equation we obtain for G has an associated error. The error for the period T, $\sigma_T$ and the error for $\alpha$, $\sigma_{\alpha}$ are obtained from the models we construct for G. Additional sigma terms involved in our final error propagation include $\sigma_d$, $\sigma_l$, $\sigma_H$, $\sigma_L$ and $\sigma_M$. Our initial equation for the erorr propagation (for a function f=G) takes the form: 

$$\sigma_{G} = \sqrt{ \left( \frac{\partial{g}}{\partial{T}} \right)^2 \sigma_{T}^2 + \left( \frac{\partial{g}}{\partial{\alpha}} \right)^2 \sigma_{\alpha}^2 + \left( \frac{\partial{g}}{\partial{d}} \right)^2 \sigma_{d}^2 + \left( \frac{\partial{g}}{\partial{l}} \right)^2 \sigma_{l}^2 + \left( \frac{\partial{g}}{\partial{M}} \right)^2 \sigma_{M}^2 + \left( \frac{\partial{g}}{\partial{L}} \right)^2 \sigma_{L}^2 + \left( \frac{\partial{g}}{\partial{H}} \right)^2 \sigma_{H}^2}$$

let $$j=\frac{\partial{g}}{\partial{T}}$$, $$k = \frac{\partial{g}}{\partial{\alpha}} $$, $$l=\frac{\partial{g}}{\partial{d}}$$,$$ m=\frac{\partial{g}}{\partial{l}}$$, $$n =\frac{\partial{g}}{\partial{M}}$$,$$ o= \frac{\partial{g}}{\partial{L}}$$, $$p = \frac{\partial{g}}{\partial{H}}$$. 

Taking, derivatives using Wolfram Mathematica, we obtain: 

$$j = \frac{ -2dl(H-\frac{dl}{4L})^2 \pi^2}{LMT^3}$$

$$k = \frac{ dl(H-\frac{dl}{4L})^2 \alpha}{2LM}$$

$$l = \frac{ -dl^2(H-\frac{dl}{4L}) (\frac{4 \pi^2}{T^2} + \alpha^2)}{8 L^2 M} + \frac{l(H-\frac{dl}{4L})^2 (\frac{4 \pi^2}{T^2} + \alpha^2)}{4 L M} $$

$$ m = \frac{ -d^2 l(H-\frac{dl}{4L}) (\frac{4 \pi^2}{T^2} + \alpha^2)}{8 L^2 M} + \frac{d(H-\frac{dl}{4L})^2 (\frac{4 \pi^2}{T^2} + \alpha^2)}{4 L M} $$

$$ n = \frac{ -d l (H-\frac{dl}{4L})^2 (\frac{4 \pi^2}{T^2} + \alpha^2)}{4 L M^2} $$

$$ o = \frac{ d^2 l^2 (H-\frac{dl}{4L}) (\frac{4 \pi^2}{T^2} + \alpha^2)}{8 L^3 M} - \frac{dl(H-\frac{dl}{4L})^2 (\frac{4 \pi^2}{T^2} + \alpha^2)}{4 L^2 M} $$ 

$$p = \frac{ d l (H-\frac{dl}{4L}) (\frac{4 \pi^2}{T^2} + \alpha^2)}{2 L M}$$  

and our $\sigma_G$ equation becomes: 

$$\sigma_{G} = \sqrt{ \left( j \right)^2 \sigma_{T}^2 + \left( k \right)^2 \sigma_{\alpha}^2 + \left( l \right)^2 \sigma_{d}^2 + \left( m \right)^2 \sigma_{l}^2 + \left( n \right)^2 \sigma_{M}^2 + \left( o \right)^2 \sigma_{L}^2 + 
\left( p \right)^2 \sigma_{H}^2}
$$ 







```python
def G_uncertainty(T_uncertainty: float, alpha_uncertainty: float, d_uncertainty: float, l_uncertainty: float, M_uncertainty: float, L_uncertainty: float, H_uncertainty: float,
                  l: float, H: float, M: float, L: float, d_best_value: float, alpha: float, T: float) -> float: 
                  
                  repeating_component_1 = (H - (d_best_value*l/4*L))  
                  repeating_component_2 = (((4*(np.pi)**2)/T**2) + alpha**2) 
                  repeating_component_3 = 4*L*M  
                  repeating_component_4 = 8*(L**2)*M 
                  repeating_component_5 = 2*L*M 

                  j_num1 = -2*d_best_value*l*(repeating_component_1**2)*(np.pi**2)  
                  j_denom1 = L*M*(T**3)  
                  j_full = j_num1 / j_denom1   
                  print(f"j_full is {j_full}")

                  k_num1 = d_best_value*l*(repeating_component_1**2)*alpha
                  k_denom1 = repeating_component_5 
                  k_full = k_num1 / k_denom1 
                  print(f"k_full is {k_full}")

                  l_num1 = -d_best_value*(l**2)*(repeating_component_1)*(repeating_component_2) 
                  l_denom1 = repeating_component_4 
                  l_num2 = l*(repeating_component_1**2)*repeating_component_2 
                  l_denom2 = repeating_component_3 
                  l_full = (l_num1/l_denom1) + (l_num2/l_denom2) 
                  print(f"l_full is {l_full}")

                  m_num1 = -(d_best_value**2)*l*(repeating_component_1)*(repeating_component_2)
                  m_denom1 = repeating_component_4 
                  m_num2 = d_best_value*(repeating_component_1**2)*(repeating_component_2) 
                  m_denom2 = repeating_component_3 
                  m_full = (m_num1/m_denom1) + (m_num2/m_denom2)
                  print(f"m_full is {m_full}")
 
                  n_num1 = -d_best_value*l*(repeating_component_1**2)*(repeating_component_2) 
                  n_denom1 = repeating_component_3*M  
                  n_full = n_num1 / n_denom1 
                  print(f"n_full is {n_full}")

                  o_num1 = (d_best_value**2)*(l**2)*repeating_component_1*repeating_component_2
                  o_denom1 = repeating_component_4*L 
                  o_num2 = d_best_value*l*(repeating_component_1**2)*repeating_component_2
                  o_denom2 = repeating_component_3*L 
                  o_full = (o_num1/o_denom1) - (o_num2/o_denom2)  
                  print(f"o_full is {o_full}")

                  p_num1 = d_best_value*l*repeating_component_1*repeating_component_2
                  p_denom1 = repeating_component_5
                  p_full = p_num1/p_denom1 
                  print(f"p_full is {p_full}")
                  
                  sigma_g = np.sqrt(((j_full**2)*(T_uncertainty**2)) + ((k_full**2)*(alpha_uncertainty**2))+
                                    ((l_full**2)*(d_uncertainty**2)) + ((m_full**2)*(l_uncertainty**2))+
                                    ((n_full**2)*(M_uncertainty**2)) + ((o_full**2)*(L_uncertainty**2))+
                                    ((p_full**2)*(H_uncertainty**2)))          
                  return sigma_g
                  
print("-----------------------------Model 1--------------------------------------")
G_uncertainty_first_df = G_uncertainty(model_1_sigma_T, model_1_alpha_uncertainty, d_uncertainty, l_uncertainty, M_uncertainty,
                                       L_uncertainty, H_uncertainty, l, H, M, L, d_best_value, model_1_alpha, model_1_T)  
print(f"G_uncertainty_first_df is {G_uncertainty_first_df}") 

print("-----------------------------Model 2--------------------------------------")
G_uncertainty_second_df = G_uncertainty(model_2_sigma_T, model_2_alpha_uncertainty, d_uncertainty, l_uncertainty, M_uncertainty,
                                       L_uncertainty, H_uncertainty, l, H, M, L, d_best_value, model_2_alpha, model_2_T)  
print(f"G_uncertainty_second_df is {G_uncertainty_second_df}")

print("-----------------------------Model 3--------------------------------------")
G_uncertainty_third_df = G_uncertainty(model_3_sigma_T, model_3_alpha_uncertainty, d_uncertainty, l_uncertainty, M_uncertainty,
                                       L_uncertainty, H_uncertainty, l, H, M, L, d_best_value, model_3_alpha, model_3_T)  
print(f"G_uncertainty_third_df is {G_uncertainty_third_df}")
```
